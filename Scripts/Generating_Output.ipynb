```python
from pyspark.sql.functions import col
import json

# Query
not_submitted = spark.read.table("Gold.dbo.gold_not_submitted_assignment_001")

# API-style JSON
response_1 = {
    "status": "success",
    "message": "Students who have not submitted Assignment_101",
    "data": [row["student_id"] for row in not_submitted.collect()]
}

submission_rate = spark.read.table("Gold.dbo.gold_submission_rate_per_course") \
    .select("course_id", "submission_rate_percent")

# API-style JSON
response_2 = {
    "status": "success",
    "message": "Submission rate per course (%)",
    "data": [
        {
            "course_id": row["course_id"],
            "submission_rate_percent": round(row["submission_rate_percent"], 2)
        }
        for row in submission_rate.collect()
    ]
}
# print(json.dumps(response_1, indent=2))
overdue_counts = spark.read.table("Gold.dbo.gold_overdue_assignments_per_student") \
    .select("student_id", "overdue_assignment_count")

# API-style JSON
response_3 = {
    "status": "success",
    "message": "Overdue assignment count per student",
    "data": [
        {
            "student_id": row["student_id"],
            "overdue_assignment_count": row["overdue_assignment_count"]
        }
        for row in overdue_counts.collect()
    ]
}


def export_results_to_json():
    result = {
        "status": "success",
        "message": "Pipeline executed and results collected.",
        "data": {
            "not_submitted_assignment_101": response_1,
            "submission_rate_per_course": response_2,
            "overdue_assignment_count_per_student": response_3
        }
    }

    return result





print(export_results_to_json())
print("âœ… Pipeline complete!")
```
